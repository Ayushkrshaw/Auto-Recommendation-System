```python
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
nlp=spacy.load("en_core_web_sm")
```

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import nltk
nltk.download('punkt')

# Initialize the PorterStemmer
stemmer = PorterStemmer()

```

```python
def clean_and_extract_tags(text):
  doc=nlp(text.lower())
  tags=[stemmer.stem(token.text) for token in doc if token.text.isalnum() and token.text not in STOP_WORDS]
  return ', '.join(tags)

columns_to_extract_tag_from = ['Category','Brand','Description','Name','Contents']



for column in columns_to_extract_tag_from:
  train[column]=train[column].apply(clean_and_extract_tags)
```

```python
def content_based_recommendations(train,item_name,top_n=10):
  if item_name not in train['Name'].values:
    print(f"Item '{item_name}' not found in the training data.")
    return pd.DataFrame()

  tfidf_vectorizer=TfidfVectorizer(stop_words='english')
  tfidf_matrix_content=tfidf_vectorizer.fit_transform(train['Tags'])
  cosine_similarities_content=cosine_similarity(tfidf_matrix_content,tfidf_matrix_content)
  item_index=train[train['Name']==item_name].index[0]
  similar_item=list(enumerate(cosine_similarities_content[item_index]))
  similar_items=sorted(similar_item,key=lambda x:x[1], reverse=True)
  top_similar_items=similar_items[1:top_n+1]
  recommended_items_index=[x[0] for x in top_similar_items]
  recommended_items_details = train.iloc[recommended_items_index][['Name','Brand','ImageURL','Description','Contents','Rating']]
  return recommended_items_details
```

```python
item_name= 'alisha, solid, women, cycling, shorts'
content_based_rec = content_based_recommendations(train,item_name,top_n=8)
content_based_rec
```

```python

def collaborative_filtering_recommendations(train,target_user_id,top_n=10):
  user_item_matrix=train.pivot_table(index='ID',columns='ProdID',values='Rating',aggfunc='mean').fillna(0).astype(int)
  user_similarity=cosine_similarity(user_item_matrix)
  target_user_index= user_item_matrix.index.get_loc(target_user_id)
  user_similarities=user_similarity[target_user_index]
  similar_user_indices=user_similarities.argsort()[::-1][1:]

  recommend_items=[]

  for user_index in similar_user_indices:
    rated_by_similar_user = user_item_matrix.iloc[user_index]
    not_rated_by_target_user = (rated_by_similar_user==0) & (user_item_matrix.iloc[target_user_index]==0)
    recommend_items.extend(user_item_matrix.columns[not_rated_by_target_user][:top_n])
  recommend_items_details=train[train['ProdID'].isin(recommend_items)][['Name','Brand','ImageURL','Description','Contents','Rating']]
  return recommend_items_details


target_user_id=4
top_n=5
collaborative_filtering_rec = collaborative_filtering_recommendations(train,target_user_id)
print(f"Top {top_n} recommendations for User {target_user_id}:")
collaborative_filtering_rec.head(5)
```

```python
def hybrid_recommendations(train,target_user_id,item_name,top_n=10):
  content_based_rec=content_based_recommendations(train,item_name,top_n)
  collaborative_filtering_rec=collaborative_filtering_recommendations(train,target_user_id,top_n)
  hybrid_rec=pd.concat([content_based_rec,collaborative_filtering_rec]).drop_duplicates()
  return hybrid_rec.head(10)
```

```python
target_user_id=2
item_name='eternal gandhi super series crystal paper weights silver finish'
hybrid_rec= hybrid_recommendations(train,target_user_id,item_name,top_n=10)
print(f"Top 10 Hybrid Recommendations for User {target_user_id} and Item '{item_name}':")
hybrid_rec
```
