```python
feat=df1['Contents'].unique()
feat=list(feat)[:100]

features=[]
for fe in feat:
  for feature in fe.split(' '):
    if not (feature.isdigit() and len(feature)>3 ):
      features.append(feature[:-1])
print(features)
```

```python
cate=df1['Category'].unique()
cate=list(cate)[:100]
types=[]
for tp in cate:
  for type in tp.split(' '):
    if not (type.isdigit() ):
      types.append(type[:-1])
print(len(types))
```

```python
import random

# Function to generate similar product names
def generate_product_names(base_name, count=10):
    generated_names = set()

    # Split base name into components
    components = base_name.split()

    while len(generated_names) < count:
        # Randomly replace components with synonyms or similar terms
        brand = random.choice(brands)
        type_ = random.choice(types)
        feature = random.choice(features)

        # Create variations
        variation1 = f"{brand} {type_} "
        variation2 = f"{brand} {components[-1]} {feature} "
        variation3 = f"{brand} {type_} {feature} {components[-1]}"

        generated_names.update([variation1, variation2, variation3])

    return list(generated_names)

# Generate 2000 product names
all_generated_names = []
for base_name in df1['Name']:
    if len(base_name)>1:
      all_generated_names.extend(generate_product_names(base_name, count=100))

# Deduplicate and limit to 2000 names
all_generated_names = list(set(all_generated_names))[:2000]

# Display generated product names
for name in all_generated_names:
    print(name)
```

```python
def reshape_array(input_array, d):
    k, _ = input_array.shape
    new_array = np.zeros((k, d))

    for i in range(k):
        for j in range(d):
            start_idx = j * (768 // d)
            end_idx = (j + 1) * (768 // d) if j < (d - 1) else 768
            chunk = input_array[i, start_idx:end_idx]
            new_array[i, j] = np.mean(chunk)

    return new_array
```

```python
step = 1000
query_dim = 32

cols = ['q' + str(x) for x in list(range(0, query_dim))] + ['query']
cnt = 0

for i in range(0,len(queries),step):

    cnt += 1
    # Define a list of sentences
    sentences = list(queries)[i:i+step]

    sentence_embeddings = model.encode(sentences, convert_to_tensor=True)
    sentence_embeddings = reshape_array(np.array(sentence_embeddings), query_dim)

    df_tmp = pd.DataFrame(np.concatenate((sentence_embeddings, np.array(sentences).reshape(-1,1)), axis=1))

    df_tmp.columns = cols

    df_tmp.to_csv(
        f'query_{cnt}.csv', header = True, index = False)

#     if cnt == 2:
#         break

    print(i)
```

```python
query_tower_input_dim = 32
product_tower_input_dim = (32*5)
```

**Finding Similar** **Queries**

```python
from annoy import AnnoyIndex
q = AnnoyIndex(query_tower_input_dim, 'euclidean')
mp_query_dict = {}

for ix,row in df_query_embedding.iterrows():
    mp_query_dict[row['qid']] = row['query']

    key = int(row['qid'])
    vec = list(row[['q'+str(x) for x in list(range(query_tower_input_dim))]])

#     print(key,vec)
    q.add_item(key,vec)
```

```python
q.build(100) # 100 trees
q.save('query.tree')


q = AnnoyIndex(query_tower_input_dim,  'euclidean')
q.load('query.tree')
```

```python
top_k = 20
mat = []
for ix,row in df_query_embedding.iterrows():
    item = row['query']
    mat.append([item] + [mp_query_dict[x] for x in q.get_nns_by_item(row['qid'], top_k+1)[1:]])

    if ix == 50:
        break

cols = ['query_id']
for i in range(top_k):
    cols += ['nearest_{}'.format(i+1)]

print(cols)

df_neighbors1 = pd.DataFrame(mat, columns = cols)

display(df_neighbors1.head(50)
       )
```

**Finding Similar Products**

```python
p = AnnoyIndex(product_tower_input_dim, 'euclidean')
mp_product_dict = {}

for ix,row in df_product_embedding.iterrows():
    mp_product_dict[int(row['pid'])] = row['Name']

    key = int(row['pid'])
    vec = list(row[['p'+str(x) for x in list(range(product_tower_input_dim))]])

#     print(key,vec)
    p.add_item(key,vec)
```

```python
p.build(100) # 100 trees
p.save('product.tree')


p = AnnoyIndex(product_tower_input_dim,  'euclidean')
p.load('product.tree')
```

```python
top_k = 20
mat = []
for ix,row in df_product_embedding.iterrows():
    item = row['Name']
    mat.append([item] + [mp_product_dict[x] for x in p.get_nns_by_item(row['pid'], top_k+1)[1:]])

    if ix == 50:
        break

cols = ['ProdID']
for i in range(top_k):
    cols += ['nearest_{}'.format(i+1)]

print(cols)

df_neighbors2 = pd.DataFrame(mat, columns = cols)

display(df_neighbors2.head(50))
```
